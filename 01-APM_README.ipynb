# Advanced Python Modules(APM) – Puzzle Solution

This notebook contains both my rough workthrough and my final solution to the **Advanced Python Modules Puzzle**  
from the Zero-to-Hero Python Bootcamp.

The objective:
- Unzip a hidden folder  
- Walk through directory structure  
- Search every file for a secret phone number  
- Use `os`, `re`, and file handling to build a real solution  
- Print the discovered phone number

---

## Skills Demonstrated
- `os.walk()` for recursive directory traversal  
- Regex pattern matching (`re.findall`)  
- File reading and path handling  
- Clean and modular Python functions  
- Real-world debugging and problem solving

## How to Run This Project
1.	Download / clone the repository
2.	Ensure you have Python 3 installed
3.	Open the notebook APM_Solution.ipynb
4.	Make sure the folder extracted_content/ is in the same directory as the notebook
5.	Run all cells from top to bottom
6.	The discovered phone number will print in the output section
